```python
#11
def func2_np(n):
  return np.sinh((1/n*(np.sum(np.random.rand(n)))))
```

```python
def func3_np(n):
  return 1/n*(np.sinh(np.random.rand(n)).sum())
 
```

```python
def func4_np(n):
  return 1/(n**2)*(np.random.rand(n,n).sum())
```
